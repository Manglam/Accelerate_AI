```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

## Q1) Which is the busiest week of the day?

```python
# since data has day variable, hence lets plot the number of days for each day in data to find the businest week of the day
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="day", data=tips)
```

From the plot above, we can say that the busiest week of the day is Saturday

# Q2) How does the total bill vary by sex?

```python
# Plot a boxplot to understand the relationship between sex and total bill
ax = sns.boxplot(x='sex', y='total_bill', data=tips)
```

From the plot above, we can say that the Male are spending more then Female

# Q3) Plot a box plot to see how Tip amount vary by day of the week?

```python
ax = sns.boxplot(x='day', y='tip', data=tips)
```

On average, tips generally on lower side on Thursday, and on higher side on Sunday
Also, Saturday seems to have a lot of outliers, so it could be a lucky day for waiters

# Q4) Draw a plot to find out of smokers pay more or less than non-smokers on Weekends? How about Friday?

```python
g = sns.FacetGrid(tips, col="day", col_order=["Thur", "Fri", "Sat", "Sun"])
ax=g.map(sns.boxplot, "smoker", "total_bill", order=["Yes", "No"])
```

Smokers are generally paying more than non-smokers, on weekends.<br>
On Friday, smokers are paying less than non-smokers.

# Q5) Draw an appropriate plot to find out, between Male and Female, who tips higher on Dinner ? Is this any different during Lunch?

```python
g = sns.FacetGrid(tips, col="time", sharey=False)
ax=g.map(sns.boxplot, "sex", "tip", order=["Male", "Female"])
```

Male and Female, on average tip similary(if we compare the median) for dinner. Male, generally have higher variation on tipping in dinner.<br>
During Lunch, Male tip higher than Female.
